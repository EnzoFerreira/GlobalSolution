# Preparando os dados e pré-processando o treinamento

## Imports

```python
import tensorflow as tf
from tensorflow.keras import models, layers
import matplotlib.pyplot as plt
import numpy as np
```

 ## Pré definições de alguns parametros

```python
TAMANHO_IMG = 256
TAMANHO_BATCH_IMG = 32
RGB_COLORS = 3
N_ITERACAO = 5
input_dados = (TAMANHO_BATCH_IMG, TAMANHO_IMG, TAMANHO_IMG, RGB_COLORS)
n_classes = 3
```

## Carregando o dataset e "printando exemplos"

```python
#Implementando o dataset
ds = tf.keras.preprocessing.image_dataset_from_directory(
    "PlantVillage",
    shuffle=True,
    image_size=(TAMANHO_IMG,TAMANHO_IMG),
    batch_size = TAMANHO_BATCH_IMG
)
```

```python
#Nomes originais das classes do dataset
class_names = ds.class_names
class_names
```

```python
#Alterando os nomes
class_names = ['Potato___Early_blight', 'Potato___Late_blight', 'Potato___healthy']
class_names_novo = ['Fungo Alternaria solani', 'Fungo Phytophthora infestans', 'Batata Saudável']
```

```python
class_names_map = {class_names[i]: class_names_novo[i] for i in range(len(class_names))}
```

```python
class_names = [class_names_map[name] for name in class_names]
print(class_names)
```

```python
#Imprimindo um exemplo de imagem contida no dataset
for image_batch, label_batch in ds.take(1):
    for i in range(1):
        plt.figure(figsize=(5,5))
        plt.imshow(image_batch[0].numpy().astype("uint8"))
        plt.title(class_names[label_batch[0]])
        plt.axis("off")
```

 ## Função para separação dos dados e configuração para o treinamento dos dados

```python
def dividindo_e_separando_dataset(ds, train_split=0.8, val_split=0.1, test_split=0.1, shuffle=True, shuffle_size=10000):
    assert (train_split + test_split + val_split) == 1
    
    ds_size = len(ds)
    
    if shuffle:
        ds = ds.shuffle(shuffle_size, seed=12)
    
    train_size = int(train_split * ds_size)
    val_size = int(val_split * ds_size)
    
    train_ds = ds.take(train_size)    
    val_ds = ds.skip(train_size).take(val_size)
    test_ds = ds.skip(train_size).skip(val_size)
    
    return train_ds, val_ds, test_ds
```

```python
#retirando os datasets separados da função acima
train_ds, val_ds, test_ds = dividindo_e_separando_dataset(ds)
```

```python
#dataset é pesado, ent o buffer ajuda no desempenho da máquina
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
#alterando o tamanho das images
resize_and_rescale = tf.keras.Sequential([
    layers.experimental.preprocessing.Resizing(TAMANHO_IMG, TAMANHO_IMG),
    layers.experimental.preprocessing.Rescaling(1.0/255)
])
```

```python
#corrigindo posições das imagens
data_augmentation = tf.keras.Sequential([
    layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical"),
    layers.experimental.preprocessing.RandomRotation(0.2),
])
```

 # Construindo o modelo de treinamento 

```python
#Construindo a rede neural
model = models.Sequential([
    resize_and_rescale,
    data_augmentation,
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64,  kernel_size = (3,3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64,  kernel_size = (3,3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64,activation='relu'),
    layers.Dense(n_classes, activation='softmax')
])
model.build(input_shape=input_dados)
```

```python
#Traz um resumo dos tipos de dados que estamos utitilzando
model.summary()
```

```python
#Utilianzdo um otimizador para ajustar os pesos durante o treinamento
model.compile(
    optimizer = 'adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=['accuracy']
)
```

```python
#Realizando o treinamento
history = model.fit(
    train_ds,
    epochs=N_ITERACAO,
    batch_size = TAMANHO_BATCH_IMG,
    verbose = 1,
    validation_data = val_ds
)
```
